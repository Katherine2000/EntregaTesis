```python
#Verificar que en la matriz de correlación no hayan columnas que estén altamente correlacionadas, en este caso
#columnas que superen el umbral de 0.85 como medida de correlación.

def correlation(dataset, threshold): 
    corr_matrix = dataset.corr()
    colum1 = []
    colum2 = []
    numCorr = []
    for i in range(len(corr_matrix.columns)):
        #print('i', i)
        for j in range(i):
            #print('j',j)
            if (corr_matrix.iloc[i, j] >= threshold):
                colum1.append(corr_matrix.columns[i])
                colum2.append(corr_matrix.columns[j])
                numCorr.append(corr_matrix.iloc[i, j])
    
    altaCorrelacion = pd.DataFrame(list(zip(colum1,colum2,numCorr)), columns = ['Columna 1','Columna 2', 'Correlación'])
    return altaCorrelacion

tabla = correlation(data,0.85)
tabla
```

-   ¿Cómo se van a manejar los datos atípicos?
